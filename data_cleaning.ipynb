```python
def laptime_conversion(t_time):
    if pd.isna(t_time) or t_time == "\\N":
        return None
    try:
        minutes, seconds = t_time.split(":")
        return int(minutes) * 60 + float(seconds)
    except:
        return None

def qualifying_time_conversion(df):
    for col in ['q1', 'q2', 'q3']:
        df[col] = df[col].apply(laptime_conversion)
    return df

def fill_missing_qual_times(row):

    if pd.isna(row['q1']) and not pd.isna(row['q2']):
        row['q1'] = row['q2']
    if pd.isna(row['q1']) and not pd.isna(row['q3']):
        row['q1'] = row['q3']

    if pd.isna(row['q2']):
        row['q2'] = row['q1']
    if pd.isna(row['q3']):
        row['q3'] = row['q2']
    
    return row


exp = qualifying_time_conversion(new_df)
exp = exp.apply(fill_missing_qual_times, axis=1)
```

1st action : taking mean where the nan values are for avg circuit relation


```python

for col in ['driver_circuit_avg_result_point_before_race', 'driver_circuit_avg_pitstops_before_race']:
    exp[col] =exp.groupby(['driverRef', 'circuitId'])[col].transform(
        lambda x: x.fillna(x.mean())
    )
exp['driver_circuit_avg_result_point_before_race'].fillna(0, inplace=True)

exp["driver_circuit_avg_pitstops_before_race"].fillna(0, inplace=True)
```

```python
def fill_with_prior_mean(df, col, group_col='driverRef'):
    
    filled_col = df[col].copy()

    # For each driver group
    for driver, group_indices in df.groupby(group_col).groups.items():
        # Get all values for the driver
        values = df.loc[group_indices, col]

        # Calculate expanding mean shifted by 1 (exclude current row)
        prior_means = values.expanding().mean().shift(1)

        # Fill NaNs in the original column with prior means
        # Only fill where original is NaN and prior mean exists
        fill_locs = values.isna() & prior_means.notna()
        filled_col.loc[group_indices[fill_locs]] = prior_means.loc[fill_locs]

    # For any remaining NaNs (like first rows), fill with overall mean of column
    filled_col.fillna(df[col].mean(), inplace=True)

    return filled_col

def fill_driver_avg_finish_last_3(df, driver_col='driverRef', finish_col='driver_avg_finish_last_3'):
    filled_col = df[finish_col].copy()

    # Group by driver
    for driver, group_indices in df.groupby(driver_col).groups.items():
        # Extract the finish positions for this driver
        driver_finish = df.loc[group_indices, finish_col]

        # Calculate rolling mean of last 3 races, shifted by 1 to exclude current race
        rolling_mean_3 = driver_finish.rolling(window=3, min_periods=1).mean().shift(1)

        # Fill NaNs in original finish_col with rolling means
        fill_locs = driver_finish.isna() & rolling_mean_3.notna()
        filled_col.loc[group_indices[fill_locs]] = rolling_mean_3.loc[fill_locs]

    # Fill any remaining NaNs with overall mean
    filled_col.fillna(df[finish_col].mean(), inplace=True)

    return filled_col


exp['driver_avg_finish_last_3'] = fill_driver_avg_finish_last_3(new_df)

exp['driver_podium_rate_prior'] = fill_with_prior_mean(new_df, 'driver_win_rate_prior')
exp['driver_avg_finish_pos_prior'] = fill_with_prior_mean(new_df, 'driver_avg_finish_pos_prior')
exp['driver_podium_rate_prior'] = fill_with_prior_mean(new_df, 'driver_podium_rate_prior')
```
